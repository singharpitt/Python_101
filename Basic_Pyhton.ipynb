## Basic Operations (Arithmetic, Comparison, Logical)

```python
# 1. Arithmetic Operations

# Define two numbers
a = 10
b = 3

# Addition
sum_result = a + b
print("Addition:", a, "+", b, "=", sum_result)  # Output: 10 + 3 = 13

# Subtraction
difference = a - b
print("Subtraction:", a, "-", b, "=", difference)  # Output: 10 - 3 = 7

# Multiplication
product = a * b
print("Multiplication:", a, "*", b, "=", product)  # Output: 10 * 3 = 30

# Division
quotient = a / b
print("Division:", a, "/", b, "=", quotient)  # Output: 10 / 3 = 3.3333...

# Floor Division (Division without the remainder)
floor_div = a // b
print("Floor Division:", a, "//", b, "=", floor_div)  # Output: 10 // 3 = 3

# Modulus (Remainder of the division)
remainder = a % b
print("Modulus:", a, "%", b, "=", remainder)  # Output: 10 % 3 = 1

# Exponentiation (Power)
power = a ** b
print("Exponentiation:", a, "**", b, "=", power)  # Output: 10 ** 3 = 1000

# 2. Comparison Operations

# Equal to
is_equal = a == b
print("\nEqual to:", a, "==", b, "->", is_equal)  # Output: False

# Not equal to
is_not_equal = a != b
print("Not equal to:", a, "!=", b, "->", is_not_equal)  # Output: True

# Greater than
is_greater = a > b
print("Greater than:", a, ">", b, "->", is_greater)  # Output: True

# Less than
is_less = a < b
print("Less than:", a, "<", b, "->", is_less)  # Output: False

# Greater than or equal to
is_greater_equal = a >= b
print("Greater than or equal to:", a, ">=", b, "->", is_greater_equal)  # Output: True

# Less than or equal to
is_less_equal = a <= b
print("Less than or equal to:", a, "<=", b, "->", is_less_equal)  # Output: False

# 3. Logical Operations

# Logical AND
logical_and = (a > 5) and (b < 5)
print("\nLogical AND: (a > 5) and (b < 5) ->", logical_and)  # Output: True

# Logical OR
logical_or = (a > 5) or (b > 5)
print("Logical OR: (a > 5) or (b > 5) ->", logical_or)  # Output: True

# Logical NOT
logical_not = not (a > 5)
print("Logical NOT: not (a > 5) ->", logical_not)  # Output: False
```

## Control Flow (if-else statements)

```python
# 1. Basic if statement
age = 18

if age >= 18:
    print("You are an adult.")  # This line will execute if age is 18 or more

# 2. if-else statement
temperature = 25

if temperature > 30:
    print("It's a hot day.")
else:
    print("It's not a hot day.")  # This line will execute if temperature is 30 or less

# 3. if-elif-else statement
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")  # This block will execute because score is between 80 and 89
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: D or F")

# 4. Nested if statements
day = "Monday"
time = 14  # 24-hour format

if day == "Monday":
    if time < 12:
        print("Good morning! It's Monday.")
    else:
        print("Good afternoon! It's Monday.")
else:
    print("It's not Monday.")

# 5. if statement with multiple conditions using logical operators
is_sunny = True
is_warm = False

if is_sunny and is_warm:
    print("It's a sunny and warm day.")
elif is_sunny and not is_warm:
    print("It's sunny but not warm.")  # This block will execute because only one condition is True
elif not is_sunny and is_warm:
    print("It's warm but not sunny.")
else:
    print("It's neither sunny nor warm.")

# 6. Ternary (Conditional) Operator
# This is a shorter form of an if-else statement
is_raining = True
weather_message = "It's raining." if is_raining else "It's not raining."
print(weather_message)  # Output: It's raining.

# 7. Checking for multiple values with 'in'
day_of_week = "Saturday"

if day_of_week in ["Saturday", "Sunday"]:
    print("It's the weekend!")
else:
    print("It's a weekday.")

# 8. Using match-case (Python 3.10 and later) - a more readable way to handle multiple conditions
grade = 'B'

match grade:
    case 'A':
        print("Excellent!")
    case 'B':
        print("Good job!")  # This block will execute because grade is 'B'
    case 'C':
        print("You passed.")
    case _:
        print("Better luck next time.")
```

## Loops (for and while)

```python
# 1. For Loop

# Example: Looping through a list of numbers
numbers = [1, 2, 3, 4, 5]

print("For Loop Example:")
for num in numbers:
    print(num)  # This will print each number in the list

# Example: Looping through a string
word = "Python"

print("\nFor Loop Through a String:")
for letter in word:
    print(letter)  # This will print each letter in the string

# Example: Using range() in a for loop
print("\nFor Loop with range():")
for i in range(5):  # range(5) generates numbers from 0 to 4
    print("Iteration:", i)

# 2. While Loop

# Example: Using while loop to count down from 5
count = 5

print("\nWhile Loop Example:")
while count > 0:
    print("Countdown:", count)
    count -= 1  # Decrement the counter

# Example: While loop with a condition
print("\nWhile Loop with Condition:")
number = 10
while number >= 0:
    if number % 2 == 0:  # Check if the number is even
        print(number, "is even")
    number -= 1  # Decrement the counter
```

## Basic Data Structures (Lists, Tuples, Dictionaries)

```python
# 1. Lists

# Creating a list
fruits = ["apple", "banana", "cherry"]

# Accessing elements by index
print("\nList Example:")
print("First fruit:", fruits[0])  # Output: apple

# Modifying a list
fruits[1] = "blueberry"  # Changing 'banana' to 'blueberry'
print("Modified list:", fruits)  # Output: ['apple', 'blueberry', 'cherry']

# Adding an element to the list
fruits.append("date")
print("After adding an element:", fruits)  # Output: ['apple', 'blueberry', 'cherry', 'date']

# Removing an element from the list
fruits.remove("cherry")
print("After removing an element:", fruits)  # Output: ['apple', 'blueberry', 'date']

# Looping through a list
print("\nLooping through a list:")
for fruit in fruits:
    print(fruit)

# 2. Tuples

# Creating a tuple
colors = ("red", "green", "blue")

# Accessing elements by index
print("\nTuple Example:")
print("First color:", colors[0])  # Output: red

# Tuples are immutable, so we can't change their elements
# colors[0] = "yellow"  # This would raise an error

# Looping through a tuple
print("Looping through a tuple:")
for color in colors:
    print(color)

# 3. Dictionaries

# Creating a dictionary
student = {
    "name": "Alice",
    "age": 20,
    "major": "Computer Science"
}

# Accessing values by keys
print("\nDictionary Example:")
print("Student Name:", student["name"])  # Output: Alice

# Modifying a value in the dictionary
student["age"] = 21
print("Updated Student Age:", student["age"])  # Output: 21

# Adding a new key-value pair
student["GPA"] = 3.8
print("After adding GPA:", student)  # Output: {'name': 'Alice', 'age': 21, 'major': 'Computer Science', 'GPA': 3.8}

# Removing a key-value pair
del student["major"]
print("After removing major:", student)  # Output: {'name': 'Alice', 'age': 21, 'GPA': 3.8}

# Looping through a dictionary
print("\nLooping through a dictionary:")
for key, value in student.items():
    print(f"{key}: {value}")
```

## Functions (Defining and Calling Functions)

```python
# 1. Defining a Basic Function
def greet(name):
    """This function greets the person whose name is passed as an argument."""
    print(f"Hello, {name}!")

# Calling the greet function with an argument
greet("Alice")  # Output: Hello, Alice!


# 2. Function with Multiple Arguments
def add_numbers(a, b):
    """This function returns the sum of two numbers."""
    return a + b

# Calling the function and storing the result
sum_result = add_numbers(5, 7)
print("Sum:", sum_result)  # Output: Sum: 12


# 3. Function with Default Arguments
def introduce(name, age=30):
    """This function introduces a person with their name and age."""
    print(f"My name is {name} and I am {age} years old.")

# Calling the function without specifying the age (uses default)
introduce("Bob")  # Output: My name is Bob and I am 30 years old.

# Calling the function with both arguments
introduce("Carol", 25)  # Output: My name is Carol and I am 25 years old.


# 4. Function with Variable-Length Arguments (*args)
def multiply(*numbers):
    """This function multiplies an arbitrary number of arguments."""
    result = 1
    for num in numbers:
        result *= num
    return result

# Calling the function with multiple arguments
product_result = multiply(2, 3, 4)
print("Product:", product_result)  # Output: Product: 24


# 5. Function with Keyword Arguments (**kwargs)
def display_info(**kwargs):
    """This function displays information provided as keyword arguments."""
    for key, value in kwargs.items():
        print(f"{key}: {value}")

# Calling the function with keyword arguments
display_info(name="Eve", age=28, job="Engineer")

```

## String Manipulation

```python
# 1. Concatenation
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print("Full Name:", full_name)  # Output: Full Name: John Doe

# 2. String Formatting
age = 25
greeting = f"Hello, my name is {full_name} and I am {age} years old."
print(greeting)  # Output: Hello, my name is John Doe and I am 25 years old.

# 3. String Methods

# Changing case
upper_name = full_name.upper()  # Converts to uppercase
print("Uppercase Name:", upper_name)  # Output: Uppercase Name: JOHN DOE

lower_name = full_name.lower()  # Converts to lowercase
print("Lowercase Name:", lower_name)  # Output: Lowercase Name: john doe

# Finding substrings
position = full_name.find("Doe")  # Finds the position of "Doe" in the string
print("Position of 'Doe':", position)  # Output: Position of 'Doe': 5

# Replacing parts of a string
replaced_name = full_name.replace("John", "Jane")
print("Replaced Name:", replaced_name)  # Output: Replaced Name: Jane Doe

# 4. String Slicing
substring = full_name[0:4]  # Extracts "John" from "John Doe"
print("Substring:", substring)  # Output: Substring: John

# 5. Splitting and Joining Strings
# Splitting a string into a list of words
words = full_name.split(" ")  # Splits "John Doe" into ["John", "Doe"]
print("Words:", words)  # Output: Words: ['John', 'Doe']

# Joining a list of words into a string
joined_name = "-".join(words)  # Joins ["John", "Doe"] into "John-Doe"
print("Joined Name:", joined_name)  # Output: Joined Name: John-Doe

# 6. Stripping Whitespace
padded_string = "   Hello, World!   "
stripped_string = padded_string.strip()  # Removes leading and trailing spaces
print("Stripped String:", stripped_string)  # Output: Stripped String: Hello, World!
```

## Input and Output (print and input)

```python
# Comprehensive Input and Output Demo

import math

def basic_input_output():
    """Demonstrates basic input and output operations."""
    name = input("Enter your name: ")
    age = input("Enter your age: ")
    print(f"Hello, {name}! You are {age} years old.")

def formatted_output():
    """Demonstrates formatted output."""
    number = 12.3456789
    print(f"Default formatting: {number}")
    print(f"Fixed-point formatting: {number:.2f}")
    print(f"Scientific notation: {number:.2e}")
    print(f"Width and alignment: |{number:10.2f}|")

def file_io_demo():
    """Demonstrates reading from and writing to a file."""
    try:
        # Writing to a file
        with open('example.txt', 'w') as file:
            file.write("Hello, World!\n")
            file.write("This is a file I/O demonstration.\n")

        # Reading from the file
        with open('example.txt', 'r') as file:
            content = file.read()
            print("\nContent of 'example.txt':")
            print(content)
    except IOError as e:
        print(f"File I/O error: {e}")

def calculate_square_root():
    """Demonstrates error handling and module usage."""
    while True:
        try:
            number = float(input("Enter a number to calculate its square root: "))
            if number < 0:
                raise ValueError("Cannot compute square root of a negative number.")
            
            sqrt_number = math.sqrt(number)
            print(f"The square root of {number} is {sqrt_number:.2f}.")
            break
        
        except ValueError as e:
            print(f"Invalid input: {e}. Please try again.")
        except Exception as e:
            print(f"An unexpected error occurred: {e}. Please try again.")

def main():
    """Main function to run all demos."""
    print("Basic Input and Output:")
    basic_input_output()
    
    print("\nFormatted Output:")
    formatted_output()
    
    print("\nFile I/O Demo:")
    file_io_demo()
    
    print("\nCalculate Square Root:")
    calculate_square_root()

# Run the main function
if __name__ == "__main__":
    main()
```

## Importing Modules

```python
# Importing Modules Demonstration

import math  # Standard library import
import numpy as np  # Importing with alias

def demonstrate_standard_module():
    """Function to demonstrate the usage of the math module."""
    print("Demonstrating math module:")
    print(f"Square root of 16: {math.sqrt(16)}")  # Using math.sqrt()
    print(f"Value of pi: {math.pi}")  # Using math.pi
    print()

def demonstrate_aliasing():
    """Function to demonstrate module aliasing."""
    print("Demonstrating aliasing with numpy:")
    array = np.array([1, 2, 3])
    print(f"Numpy array: {array}")
    print()

def demonstrate_dynamic_import():
    """Function to demonstrate dynamic import."""
    print("Demonstrating dynamic import:")
    module_name = "math"
    math_module = __import__(module_name)
    print(f"Square root of 25 using dynamic import: {math_module.sqrt(25)}")
    print()

def demonstrate_import_path():
    """Function to demonstrate the module search path."""
    import sys
    print("Module search path:")
    for path in sys.path:
        print(path)
    print()

# Running all demonstrations
if __name__ == "__main__":
    demonstrate_standard_module()
    demonstrate_aliasing()
    demonstrate_dynamic_import()
    demonstrate_import_path()
```
